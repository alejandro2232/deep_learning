#Ejercicio 01 producto + iva



```python
prod1,prod2,prod3,prod4,prod5= input("input product ").split(",");
suma = int(prod1) + int(prod2) + int(prod3) + int(prod4) + int(prod5);
print()
iva= int(suma) * 0.19;
print()
valorTotal = int(suma) + iva;
print("Suma: ",suma,"   Iva: ",iva, "   Valor total: ", valorTotal);


```
